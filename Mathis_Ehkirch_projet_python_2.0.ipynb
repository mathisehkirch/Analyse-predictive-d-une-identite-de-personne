```python
# Obtenez les colonnes manquantes dans X_train
missing_columns_train = set(X_test.columns) - set(X_train.columns)

# Obtenez les colonnes manquantes dans X_test
missing_columns_test = set(X_train.columns) - set(X_test.columns)

# Ajoutez les colonnes manquantes  X_train avec des valeurs par dfaut
for column in missing_columns_train:
    X_train[column] = 0  # ou toute autre valeur par dfaut

# Ajoutez les colonnes manquantes  X_test avec des valeurs par dfaut
for column in missing_columns_test:
    X_test[column] = 0  # ou toute autre valeur par dfaut
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

kmeans_train=KMeans(n_clusters=32, init='k-means++', n_init="auto", max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='lloyd').fit_predict(X_train)
kmeans_train
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

kmeans_test=KMeans(n_clusters=32, init='k-means++', n_init="auto", max_iter=300, tol=0.0001, verbose=0, random_state=None, copy_x=True, algorithm='lloyd').fit_predict(X_test)
kmeans_test
```

```python
#X_train = pd.DataFrame(X_train)
#X_test = pd.DataFrame(X_test)
# Crer un DataFrame  partir de la liste kmeans
kmeans_df1 = pd.DataFrame({'cluster': kmeans_train})
kmeans_df2 = pd.DataFrame({'cluster': kmeans_test})

# Ajouter la colonne 'cluster'  X
X_train = X_train.assign(cluster=kmeans_df1)
X_test = X_test.assign(cluster=kmeans_df2)


# Afficher les donnes avec les clusters assigns
print(X_train.head())
print(X_test.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import f1_score

#X_train=data2.drop("Player",axis=1).values
#Y_train=data2["Player"].values
#X_test=data1.values

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

neigh = KNeighborsClassifier(n_neighbors=10)


neigh.fit(X_train,Y_train)
y_pred=neigh.predict(X_test)
```

```python
# Crer un DataFrame avec les ID et les prdictions
n_samples = len(y_pred)
reference_ids = range(1,n_samples+1)
submission_df = pd.DataFrame({
    'RowId':reference_ids ,
    'prediction': y_pred
})

# Enregistrer le DataFrame au format CSV
submission_df.to_csv('submission.csv', index=False)
```

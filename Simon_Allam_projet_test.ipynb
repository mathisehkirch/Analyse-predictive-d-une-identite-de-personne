```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = pd.read_csv("TEST.CSV",delimiter=";",header=None)
data2 = pd.read_csv("TRAIN.CSV",delimiter=";",header=None)
```

```python
#Création de 2 colonnes, la colonne comprenant la race et la colonne comprenant les touches
data1["Race"]=data1.iloc[:,0].str.split(',').str[0]
data1["move"]=data1.iloc[:,0].str.split(',').str[1:]
data1.drop(data1.columns[0],axis=1,inplace=True)
data1.head()
```

```python
data1.info()
```

```python
#Création de 3 colonnes, la colonne comprenant le nom du joueur, la colonne comprenant la race et la colonne comprenant les touches

data2["Player"]=data2.iloc[:,0].str.split(',').str[0]
data2["Race"]=data2.iloc[:,0].str.split(',').str[1]
data2["move"]=data2.iloc[:,0].str.split(',').str[2:]
data2.drop(data2.columns[0],axis=1,inplace=True)
data2.head()
```

```python
#Création d'une colonne par touches avec le compte des touches à chaque ligne
# Étendre chaque liste dans la colonne "Move" en lignes
expanded_data2 = data2["move"].explode()

# Compter le nombre de fois où chaque classe apparaît dans chaque ligne
counts = expanded_data2.groupby(level=0).value_counts().unstack(fill_value=0)

# Renommer les colonnes avec un préfixe pour la clarté
counts.columns = [f'{col}' for col in counts.columns]

# Concaténer les données comptées avec le DataFrame d'origine
data2 = pd.concat([data2, counts], axis=1)

# Supprimer la colonne "move" d'origine
data2 = data2.drop(columns=['move'])
```

```python
#Création d'une colonne par touches avec le compte des touches à chaque ligne
# Étendre chaque liste dans la colonne "Move" en lignes
expanded_data1 = data1["move"].explode()

# Compter le nombre de fois où chaque classe apparaît dans chaque ligne
counts = expanded_data1.groupby(level=0).value_counts().unstack(fill_value=0)

# Renommer les colonnes avec un préfixe pour la clarté
counts.columns = [f'{col}' for col in counts.columns]

# Concaténer les données comptées avec le DataFrame d'origine
data1 = pd.concat([data1, counts], axis=1)

# Supprimer la colonne "move" d'origine
data1 = data1.drop(columns=['move'])
```

```python
#On créer un modèle performant à l'aide du randomforest pour booster les résultats
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import f1_score, make_scorer
from scipy.stats import randint

# RandomForestClassifier
rf_classifier = RandomForestClassifier()

# Définir les hyperparamètres à rechercher
param_dist = {
    'n_estimators': randint(100, 1000),
    'max_depth': [None] + list(randint(1, 50).rvs(10)),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

#Définir la métrique F1-Score pour une classification multiclasse
f1_scorer = make_scorer(f1_score, average='macro')

grid_search = RandomizedSearchCV(estimator=rf_classifier, param_distributions=param_dist, n_iter=25, cv=5, scoring=f1_scorer, random_state=42)

grid_search.fit(X_train, Y_train)

# Meilleurs hyperparamètres trouvés
print("Meilleurs hyperparamètres:", grid_search.best_params_)

best_rf_classifier = RandomForestClassifier(**grid_search.best_params_)

best_rf_classifier.fit(X_train, Y_train)

y_pred = best_rf_classifier.predict(X_test)
```
